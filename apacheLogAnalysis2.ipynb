```python
# imports for nlp

import spacy
import en_core_web_sm
from spacy import displacy
from spacy import matcher
```

```python
# loading en to nlp
nlp = en_core_web_sm.load()

# reading log file

logFile = open('./error1.log', 'r')
logs = logFile.read()
logFile.close()

doc = nlp(logs[:2000])

# tokenization of doc
token_list = []
for tokens in doc:
    token_list.append(tokens.text)
print(token_list)
```

```python
#from spacy.lang.en.stop_words import STOP_WORDS
stop_words = nlp.Defaults.stop_words
 
stop_words |= {'-', '/', '"', '[', ']', '(', ')'} # custom stop words appended

filtered_list = []
for token in token_list:
    if token not  in stop_words:
        filtered_list.append(token)
print(filtered_list)
```

```python
# unfiltered doc render

displacy.render(doc, style='ent',jupyter=True)
```

```python

filtered_text = ""

for word in filtered_list:
    filtered_text = filtered_text + " "+ str(word)
print(filtered_text)
```

```python
# filtered doc render

filtered_doc = nlp(filtered_text.strip().lower())

displacy.render(filtered_doc, style='ent', jupyter=True)
```

```python
ents_tokens_list = []
for token in filtered_doc.ents:
    ents_tokens_list.append(token)
print(ents_tokens_list)
```
